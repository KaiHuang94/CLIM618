### OLR regressed against RMM Index

```python
file_lr1 = "/homes/khuang8/clim680/RMM1.OLR.regression.nc"
ds_lr1 = xr.open_dataset(file_lr1)
ds_lr1
```

```python
ds_lr1=ds_lr1.reindex(lat=list(reversed(ds_lr1['lat'])))
ds_lr1
```

```python
file_lr2 = "/homes/khuang8/clim680/RMM2.OLR.regression.nc"
ds_lr2 = xr.open_dataset(file_lr2)
ds_lr2
```

```python
ds_lr2=ds_lr2.reindex(lat=list(reversed(ds_lr2['lat'])))
ds_lr2
```

### Only for OLR values above 95% signifcance level

```python
mask_c1=np.where(ds_cp1['pvalue']<0.05,ds_cp1['olr'],np.nan)
mask_c1
plt.contourf(mask_c1)
```

```python
mask_c2=np.where(ds_cp2['pvalue']<0.05,ds_cp2['olr'],np.nan)
mask_c2
plt.contourf(mask_c2)
```

```python
mask_l1=np.where(ds_lr1['pvalue']<0.05,ds_lr1['olr'],np.nan)
mask_l1
plt.contourf(mask_l1)
```

```python
mask_l2=np.where(ds_lr2['pvalue']<0.05,ds_lr2['olr'],np.nan)
mask_l2
plt.contourf(mask_l2)
```

### Difference between composite and linearly regression

```python
mask_diff1 = mask_c1-mask_l1 
mask_diff1
plt.contourf(mask_diff1)
```

```python
mask_diff2 = mask_c2-mask_l2 
mask_diff2
plt.contourf(mask_diff2)
```

```python
mask_diff=[mask_diff1,mask_diff2]
mask_diff
```

```python
raw_data = [mask_l1,mask_l2]
raw_data
```

```python
labels=['RMM1',
        'RMM2']

clevs=np.arange(-6,7,1)
rawlevs=np.arange(-16,17,4)

fig,axs=plt.subplots(nrows=2,ncols=1,
                     subplot_kw={'projection':ccrs.PlateCarree(central_longitude=180)},
                     figsize=(12,11))

axs=axs.flatten()

for i,phase in enumerate(mask_diff): 
    data=mask_diff[i]
    rawdata=raw_data[i]
    data,lons=add_cyclic_point(data,coord=ds_lr1['lon'])
    rawdata,lons=add_cyclic_point(rawdata,coord=ds_lr1['lon'])
    
    cs=axs[i].contourf(lons,ds_lr1['lat'],data,clevs, 
                       transform=ccrs.PlateCarree(),
                       cmap='BrBG_r',extend='both')
    
    axs[i].contour(lons,ds_lr1['lat'],rawdata,rawlevs, 
                       transform=ccrs.PlateCarree(),colors='red')
    
    label_latlon(axs[i],np.arange(-180,181,60),np.arange(-30,31,10))
    
    
    axs[i].set_title(labels[i])
    axs[i].coastlines()
    
fig.subplots_adjust(bottom=0.25, top=0.9, left=0.05, right=0.95,
                    wspace=0.2, hspace=0.1)

cbar_ax = fig.add_axes([0.2, 0.2, 0.6, 0.02])

cbar=fig.colorbar(cs, cax=cbar_ax,orientation='horizontal',label='W/m^2')

# Add a big title at the top
plt.suptitle('Compisited OLR - Regressed OLR')
plt.savefig('MJO_RMM1&2_diff_raw.png')
```

### Standardize the OLR to extract the structure

Since the composted OLR is much greater than the regressed OLR, the pattern of their difference is very similar to the raw composite pattern. 
Therefore, to better investigate the structure difference between composited and regressed OLR, the composited and regressed OLR is standardized by their pattern standard deviation. 

```python
nx=len(ds_cp1['lon'])
ny=len(ds_cp1['lat'])
```

```python
data=np.reshape(mask_c1,((ny*nx)))
mag = np.nanstd(data)
mag
```

```python
mask_c1 = mask_c1/mag
```

```python
data=np.reshape(mask_c2,((ny*nx)))
mag = np.nanstd(data)
mag
```

```python
mask_c2 = mask_c2/mag
```

```python
data=np.reshape(mask_l1,((ny*nx)))
mag = np.nanstd(data)
mag
```

```python
mask_l1 = mask_l1/mag
```

```python
data=np.reshape(mask_l2,((ny*nx)))
mag = np.nanstd(data)
mag
```

```python
mask_l2 = mask_l2/mag
```

```python
mask_diff=[mask_c1-mask_l1,mask_c2-mask_l2]
mask_diff
```

```python
labels=['RMM1',
        'RMM2']

clevs=np.arange(-0.5,0.51,0.01)
rawlevs=np.arange(-16,17,4)

fig,axs=plt.subplots(nrows=2,ncols=1,
                     subplot_kw={'projection':ccrs.PlateCarree(central_longitude=180)},
                     figsize=(12,11))

axs=axs.flatten()

for i,phase in enumerate(mask_diff): 
    data=mask_diff[i]
    rawdata=raw_data[i]
    data,lons=add_cyclic_point(data,coord=ds_lr1['lon'])
    rawdata,lons=add_cyclic_point(rawdata,coord=ds_lr1['lon'])
    cs=axs[i].contourf(lons,ds_lr1['lat'],data,clevs, 
                       transform=ccrs.PlateCarree(),
                       cmap='BrBG_r',extend='both')
    
    axs[i].contour(lons,ds_lr1['lat'],rawdata,rawlevs, 
                       transform=ccrs.PlateCarree(),colors='red')
    
    label_latlon(axs[i],np.arange(-180,181,60),np.arange(-30,31,10))
    
    axs[i].set_title(labels[i])
    axs[i].coastlines()
    
fig.subplots_adjust(bottom=0.25, top=0.9, left=0.05, right=0.95,
                    wspace=0.25, hspace=0.5)

cbar_ax = fig.add_axes([0.2, 0.2, 0.6, 0.02])

cbar=fig.colorbar(cs, cax=cbar_ax,orientation='horizontal',label='/std')

# Add a big title at the top
plt.suptitle('Compisited OLR - Regressed OLR')
plt.savefig('MJO_RMM1&2_diff_std.png')
```
